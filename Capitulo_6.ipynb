# **Ejemplo 30. Entrenamiento de una U-Net para la segmentación de la hoja principal del dataset RoCoLe**

````python
import os
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import layers, Model

# =========================================================
# 1. LECTURA Y PREPROCESAMIENTO DE LAS IMÁGENES
# =========================================================
# 1.1. Configuración inicial
np.random.seed(42)
tf.random.set_seed(42)

TAM = 224

originales = ["C10P11E2.jpg", "C10P10H2.jpg", "C10P10H1.jpg",
              "C10P10E2.jpg", "C10P10E1.jpg"]

editadas = ["mascara_1.png", "mascara_2.png", "mascara_3.png",
            "mascara_4.png", "mascara_5.png"]

for nombre in originales + editadas:
    if not os.path.exists(nombre):
        raise FileNotFoundError(f"No se encontró el archivo: {nombre}")


# 1.2. Lectura y normalización de las imágenes originales
X = []

for nombre in originales:
    imagen_bgr = cv2.imread(nombre)

    if imagen_bgr is None:
        raise ValueError(f"No fue posible leer la imagen: {nombre}")

    imagen_rgb = cv2.cvtColor(imagen_bgr, cv2.COLOR_BGR2RGB)
    imagen_rgb = cv2.resize(imagen_rgb, (TAM, TAM),
                            interpolation=cv2.INTER_AREA)
    imagen_rgb = imagen_rgb.astype(np.float32) / 255.0
    X.append(imagen_rgb)

X = np.array(X, dtype=np.float32)


# 1.3. Construcción de las máscaras de referencia
Y = []
for nombre in editadas:
    imagen = cv2.imread(nombre, cv2.IMREAD_UNCHANGED)
    if imagen is None:
        raise ValueError(f"No fue posible leer la imagen editada: {nombre}")
    # Caso 1: imagen PNG con transparencia
    if imagen.ndim == 3 and imagen.shape[2] == 4:
        canal_alfa = imagen[:, :, 3]
        mascara = (canal_alfa > 10).astype(np.uint8)
    # Caso 2: imagen con fondo blanco
    else:
        if imagen.ndim == 2:
            imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_GRAY2RGB)
        else:
            imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

        fondo_blanco = np.all(imagen_rgb >= 245, axis=2)
        mascara = (~fondo_blanco).astype(np.uint8)
    mascara = cv2.resize(mascara, (TAM, TAM),
                         interpolation=cv2.INTER_NEAREST)
    mascara = mascara.astype(np.float32)
    mascara = np.expand_dims(mascara, axis=-1)
    Y.append(mascara)
Y = np.array(Y, dtype=np.float32)


# 1.4. Verificación de imágenes y máscaras
print("Forma de X:", X.shape)
print("Forma de Y:", Y.shape)

for i in range(len(Y)):
    valores = np.unique(Y[i])
    porcentaje_hoja = 100 * np.mean(Y[i])
    print(f"Máscara {i + 1}: valores = {valores}, "
          f"porcentaje de hoja = {porcentaje_hoja:.2f}%")

plt.figure(figsize=(15, 6))

for i in range(5):
    plt.subplot(2, 5, i + 1)
    plt.imshow(X[i])
    plt.title(f"Original {i + 1}")
    plt.axis("off")

    plt.subplot(2, 5, i + 6)
    plt.imshow(Y[i, :, :, 0], cmap="gray", vmin=0, vmax=1)
    plt.title(f"Máscara {i + 1}")
    plt.axis("off")

plt.tight_layout()
plt.show()


# 1.5. División entre entrenamiento y validación
X_train, Y_train = X[:4], Y[:4]
X_val, Y_val = X[4:], Y[4:]

print("Entrenamiento:", X_train.shape, Y_train.shape)
print("Validación:", X_val.shape, Y_val.shape)


# =========================================================
# 2. CONSTRUCCIÓN DE LA ARQUITECTURA
# =========================================================
# 2.1. Métricas y función de pérdida
def dice_coef(y_true, y_pred, suavizado=1e-6):
    y_true = tf.reshape(tf.cast(y_true, tf.float32), [-1])
    y_pred = tf.reshape(tf.cast(y_pred, tf.float32), [-1])

    interseccion = tf.reduce_sum(y_true * y_pred)
    suma = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)

    return (2.0 * interseccion + suavizado) / (suma + suavizado)

def iou_coef(y_true, y_pred, suavizado=1e-6):
    y_true = tf.reshape(tf.cast(y_true, tf.float32), [-1])
    y_pred = tf.reshape(tf.cast(y_pred, tf.float32), [-1])

    interseccion = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred) - interseccion

    return (interseccion + suavizado) / (union + suavizado)

def perdida_bce_dice(y_true, y_pred):
    bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
    bce = tf.reduce_mean(bce)
    return bce + (1.0 - dice_coef(y_true, y_pred))


# 2.2. Bloque convolucional
def bloque_conv(x, filtros):
    x = layers.Conv2D(filtros, 3, padding="same", activation="relu",
                      kernel_initializer="he_normal")(x)

    x = layers.Conv2D(filtros, 3, padding="same", activation="relu",
                      kernel_initializer="he_normal")(x)
    return x


# 2.3. Construcción de una U-Net pequeña
entrada = layers.Input(shape=(TAM, TAM, 3))

# Codificador
c1 = bloque_conv(entrada, 16)
p1 = layers.MaxPooling2D((2, 2))(c1)
c2 = bloque_conv(p1, 32)
p2 = layers.MaxPooling2D((2, 2))(c2)
c3 = bloque_conv(p2, 64)
p3 = layers.MaxPooling2D((2, 2))(c3)

# Cuello de botella
puente = bloque_conv(p3, 128)

# Decodificador
u3 = layers.Conv2DTranspose(64, 2, strides=2, padding="same")(puente)
u3 = layers.Concatenate()([u3, c3])
c4 = bloque_conv(u3, 64)
u2 = layers.Conv2DTranspose(32, 2, strides=2, padding="same")(c4)
u2 = layers.Concatenate()([u2, c2])
c5 = bloque_conv(u2, 32)
u1 = layers.Conv2DTranspose(16, 2, strides=2, padding="same")(c5)
u1 = layers.Concatenate()([u1, c1])
c6 = bloque_conv(u1, 16)

# Salida: probabilidad de pertenecer a la hoja
salida = layers.Conv2D(1, 1, activation="sigmoid")(c6)
modelo = Model(inputs=entrada, outputs=salida, name="UNet_pequena")
modelo.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
               loss=perdida_bce_dice,
               metrics=[dice_coef, iou_coef])
modelo.summary()


# =========================================================
# 3. PROCESO DE ENTRENAMIENTO
# =========================================================
# 3.1. Configuración de callbacks
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor="val_loss", patience=25, restore_best_weights=True
    ),

    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss", factor=0.5, patience=10, min_lr=1e-6
    )
]

# 3.2. Entrenamiento de la U-Net
historial = modelo.fit(
    X_train, Y_train,
    validation_data=(X_val, Y_val),
    epochs=150,
    batch_size=1,
    callbacks=callbacks,
    verbose=1
)

# 3.3. Curvas de entrenamiento
plt.figure(figsize=(8, 4))
plt.plot(historial.history["loss"], label="Pérdida entrenamiento")
plt.plot(historial.history["val_loss"], label="Pérdida validación")

plt.xlabel("Época")
plt.ylabel("Pérdida")
plt.title("Evolución del entrenamiento")
plt.legend()
plt.tight_layout()
plt.show()


# =========================================================
# 4. EVALUACIÓN DEL MODELO
# =========================================================

# 4.1. Predicción sobre la imagen de validación
prediccion = modelo.predict(X_val, verbose=0)[0, :, :, 0]

umbral = 0.5
mascara_predicha = (prediccion >= umbral).astype(np.uint8)
segmentada = X_val[0] * mascara_predicha[:, :, np.newaxis]


# 4.2. Visualización del resultado
plt.figure(figsize=(16, 4))

plt.subplot(1, 4, 1)
plt.imshow(X_val[0])
plt.title("Imagen original")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(prediccion, cmap="gray", vmin=0, vmax=1)
plt.title("Probabilidad predicha")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(mascara_predicha, cmap="gray", vmin=0, vmax=1)
plt.title("Máscara predicha")
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(segmentada)
plt.title("Imagen segmentada")
plt.axis("off")

plt.tight_layout()
plt.show()


# 4.3. Comparación con la máscara de referencia
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(Y_val[0, :, :, 0], cmap="gray", vmin=0, vmax=1)
plt.title("Máscara de referencia")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(prediccion, cmap="gray", vmin=0, vmax=1)
plt.title("Probabilidad predicha")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(mascara_predicha, cmap="gray", vmin=0, vmax=1)
plt.title("Máscara binaria predicha")
plt.axis("off")

plt.tight_layout()
plt.show()
````
